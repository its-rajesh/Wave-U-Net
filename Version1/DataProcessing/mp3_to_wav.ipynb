# Converting .mp3 to .wav

```python
import os
```

```python
from pydub import AudioSegment
```

```python
path = '/users/rajeshr/Desktop/SaragaSorted/train/'
files = [path+i for i in os.listdir(path)]
```

```python
des_path = '/users/rajeshr/Desktop/SaragaSortedWav/train/'
```

```python
for folders in files:
    if folders != '/users/rajeshr/Desktop/SaragaSorted/train/.DS_Store':
        for file in os.listdir(folders):
            if file != '.DS_Store':
                p = folders+'/'+file
                splt = folders.split('/')[-1]
                file_ext = file.split('.')[-2]
                dest_p = des_path+splt+'/'+file_ext+'.wav'
                print(p, dest_p)

                os.makedirs(des_path+splt+'/', exist_ok = True)

                sound = AudioSegment.from_mp3(p)
                sound.export(dest_p, format="wav")
```
