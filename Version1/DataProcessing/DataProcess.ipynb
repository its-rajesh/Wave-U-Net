### Making sure only 5 files are there (mixed + 4 inst: vocal, violin, left and right)

```python
path = '/users/rajeshr/Desktop/Saraga Sorted/'
files = [path+i for i in os.listdir(path)]
```

```python
for p in files:
    if p != '/users/rajeshr/Desktop/Saraga Sorted/.DS_Store':
        for i in os.listdir(p):
            s = i.split('.')
            if s[-2] == 'multitrack-violin' or s[-2] == 'violin':
                os.rename(p+'/'+i, p+'/violin.mp3')
            elif s[-2] == 'multitrack-mridangam-left' or s[-2] == 'mridangam_left':
                os.rename(p+'/'+i, p+'/mridangam_left.mp3')
            elif s[-2] == 'multitrack-mridangam-right'or s[-2] == 'mridangam_right':
                os.rename(p+'/'+i, p+'/mridangam_right.mp3')
            elif s[-2] == 'multitrack-vocal' or s[-2] == 'vocal':
                os.rename(p+'/'+i, p+'/vocal.mp3')
            elif s[-2] == 'mp3' or s[-2] == 'mixed':
                os.rename(p+'/'+i, p+'/mixed.mp3')
            else:
                os.remove(p+'/'+i)
                print(p+'/'+i, 'REMOVED!')
            print('DONE.')
```
